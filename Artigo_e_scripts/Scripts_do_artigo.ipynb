```python
# Importando Bibliotecas
import sympy as sy
# Declarando a variavel simbolica: x
x = sy.Symbol('x')
# Declarando a Função: f(x)
f = (x**2-2*x-8)/(x-4)
# Visualizando f
print('----------------------------')
print('f =',f)
# Valor da Tendencia: a
# Valor do Limite de f para x --> a: L
a = 4
L = sy.limit(f,x,a)
# Visualizando L
print("L =",L)
print('----------------------------')
```

```python
# Importando Bibliotecas
import sympy as sy
# Declarando a variavel simbolica: x
t = sy.Symbol('t')
# Declarando a Função: F(x)
f = 10*sy.exp(-1*t)*sy.sin(2*sy.pi*t)
# Visualizando f
print('f =',f)
# Integral de f(x) em relaçao a x: Ifx
# Limites de Integração: [a,b]
a = 0
b = 1
Ifx = sy.integrate(f(t, a, b))
i_vmq = sy.sqrt(Ifx)
# Visualizando Ifx
print('Ifx =',Ifx)
# Visualizando i_vmq
print('i_vmq =',i_vmq)
```

```python
# Importando Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
# Declarando o intervalo de analise
# t = np.linspace(inicio, fim, numero de pontos)
t = np.linspace(-1,6,100)
# Declarando as Funções Receita, Custo e Lucro
R = -t**2 + 11*t
C = -t**3 + 7*t**2 - 4*t
L = R- C
# Representação Grafica das funções
# Grafica das funções R e C
plt.figure(1)
plt.plot(t,R,'m',label = 'R(t)', linewidth = 3)
plt.plot(t,C,'c',label = 'C(t)', linewidth = 3)
plt.xlabel('t (ano)')
plt.ylabel(' Receita, Custo ($)')
plt.title('Caso - III')
plt.legend()
plt.grid(True)
# Grafica da funções L
plt.figure(2)
plt.plot(t,L,'m',label = 'L(t) = R(t) - C(t)', linewidth = 3)
plt.xlabel('t (ano)')
plt.ylabel(' Lucro ($)')
plt.title('Caso - III')
plt.legend()
plt.grid(True)
```

```python
# Importando Bibliotecas
import sympy as sy
# Declarando variavel simbolica: t
t = sy.symbols('t')
print('Equação a ser resolvida:')
# Declarando Equação a ser resolvida: L = R - C
R = -t**2 + 11*t
C = -t**3 + 7*t**2 - 4*t
L = R- C
print('L =', L)
# Conjunto solução: S
S = sy.solve(L)
print('Solução da Equação')
print('S =', S)
```

```python
# Importando Bibliotecas
import sympy as sy
# Declarando a variavel simbolica: x
t = sy.symbols('t')
# Declarando as Funções: R(t) - C(t)
R = -t**2 + 11*t
C = -t**3 + 7*t**2 - 4*t
# Visualizando R e C
print('R =', R)
print('C =', C)
# Limites de Integração: [a,b]
a = 0.0
b = 5.0
# Integral de R em relação a t: IRt
IRt = sy.integrate(R,(t,a,b))
# Integral de C em relação a t: ICt
ICt = sy.integrate(C,(t,a,b))
# Arredondando: round(Variavel, nº de casas)
# Visualizando IRt
print('IRt =', round(IRt,2))
# Visualizando ICt
print('ICt =', round(ICt,2))
# Valor do lucro: L = IRt - ICt
L = IRt - ICt
# Visualizando o Valor do Lucro
print('L =',round(L,2))
```
